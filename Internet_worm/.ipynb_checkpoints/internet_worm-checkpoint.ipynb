```python
import requests
from lxml import etree
import numpy as np
import pandas as pd
```

```python
add='http://maoyan.com/board'
header={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Safari/537.36'}
```

```python
def getmess(add):
    r=requests.get(add,headers=header)
    print(r)
    return r.text
def decode(text):
    html=etree.HTML(text)
    names=html.xpath('//p[@class="name"]/a/@title')
    star=html.xpath('//p[@class="star"]/text()')
    releasetime=html.xpath('//p[@class="releasetime"]/text()')
#     print(names)
#     print(star)
#     print(releasetime)
#     print(len(names))
#     print(len(star))
#     print(len(releasetime))

    for i,j,k in zip(names,star,releasetime):
        j=j.strip(' \n')
        print(i,'\t',j,'\t',k)
    names=[i.strip("' ") for i in names]
    s1=np.array(names)
    star=[i.strip("'\n ") for i in star]
    s2=np.array(star)
    releasetime=[i.strip("' ") for i in releasetime]
    s3=np.array(releasetime)
    s=np.vstack((s1,s2,s3))
    print(s.T)
    print(s.shape)
    df=pd.DataFrame(s.T)
    return df
```

```python
test=getmess(add)
decode(test)
```

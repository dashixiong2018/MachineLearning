# 修改

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.arange(40).reshape((4,10)),index=pd.date_range('20181101',periods=4),columns=['A','B','C','D','E','F','G','H','I','J'])
print(df)
```

## 修改值

```python
df.loc['20181101',"A"]=111
df
```

```python
df.loc['20181101':'20181103',"A"]=222
df
```

```python
df.loc['20181101':'20181103',["A","D"]]=333
df
```

```python
df.iloc[:2,3:5]=444
df
```

```python
df.iloc[:2,[1,3,5]]=555
df
```

```python
df=pd.DataFrame(np.arange(40).reshape((4,10)),index=pd.date_range('20181101',periods=4),columns=['A','B','C','D','E','F','G','H','I','J'])
df[df.A>15]
```

更改选中的**所有列**

```python
df[df.A>15]=666
df
```

更改选中的**某一列**

```python
df.B[df.A>15]=777
df
```

```python
df['D'][df.A>15]=888
df
```

 添加一个属性(列)

```python
df['K']=pd.Series(np.arange(3),index=pd.date_range('20181101',periods=3))
df
```

添加一个空的列

```python
df['L']=np.nan
df
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.arange(50).reshape((5,10)),index=pd.date_range('20181101',periods=5),columns=['A','B','C','D','E','F','G','H','I','J'])
df
```

```python
df.loc['20181102']=np.nan
df.iloc[0,3]=np.nan
df.iloc[3,4]=np.nan
df
```

```python
df.dropna(axis=0,how='any')
```

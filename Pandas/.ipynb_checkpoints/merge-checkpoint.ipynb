```python
import numpy as np
import pandas as pd
```

```python
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'a':['A0','A1','A2','A3'],
                    'b':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'c':['C0','C1','C2','C3'],
                    'd':['D0','D1','D2','D3']})
print(left)
print(right)
```

```python
res=pd.merge(left,right,on='key')
res
```

```python
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'a':['A0','A1','A2','A3'],
                    'b':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':{1:'k0',2:'k1',3:'k2',4:'k3'},
                    'c':{1:'C0',2:'C1',3:'C2',4:'C3'},
                    'd':{1:'D0',2:'D1',3:'D2',4:'D3'}})
print(left)
print(right)
```

在left和right的index不对齐的情况下，结果还是和上面一样，
因为此时我们以key值作为连接的标识，不考虑其他的cloumns

```python
res=pd.merge(left,right,on='key')
res
```

在on的值不完全匹配时(只保留匹配的项)（内连接）

```python
left=pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'a':['A0','A1','A2','A3'],
                    'b':['B0','B1','B2','B3']})
right=pd.DataFrame({'key':{1:'k4',2:'k1',3:'k2',4:'k3'},
                    'c':{1:'C4',2:'C1',3:'C2',4:'C3'},
                    'd':{1:'D4',2:'D1',3:'D2',4:'D3'}})
res=pd.merge(left,right,on='key')
res
```

merge的连接默认为内连接（how='inner'），还有外链接（outer）、左外连接（left）、右外连接（right）

```python
left=pd.DataFrame({'key1':['k0','k0','k1','k2'],
                   'key2':['k0','k1','k0','k1'],
                    'a':['A0','A1','A2','A3'],
                    'b':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['k0','k1','k1','k2'],
                   'key2':['k0','k0','k0','k0'],
                    'c':['C4','C1','C2','C3'],
                    'd':['D4','D1','D2','D3']})
print(left)
print(right)
```

以两个key做连接(要求两个key都匹配才能连接)
内连接
左边有一项和右边的两项完全匹配时保留两项（左边的一项值重复用）

```python
res=pd.merge(left,right,on=['key1','key2'])
res
```

```python
left=pd.DataFrame({'key1':['k0','k0','k1','k2','k1'],
                   'key2':['k0','k1','k0','k1','k0'],
                    'a':['A0','A1','A2','A3','A4'],
                    'b':['B0','B1','B2','B3','B4']})
right=pd.DataFrame({'key1':['k0','k1','k1','k2'],
                   'key2':['k0','k0','k0','k0'],
                    'c':['C4','C1','C2','C3'],
                    'd':['D4','D1','D2','D3']})
print(left)
print(right)
```

当左边有两项和右边的两项完全匹配时，变成四项（两两组合）
同理我们可以知道当左边的m项和右边的n项完全匹配时，会产生m*n项

```python
res=pd.merge(left,right,on=['key1','key2'])
res
```

内连接时（how='inner'）不匹配项全部舍弃，外链接时（outer）则会保留，并填充默认值NaN
外链接在内连接的基础上多了未匹配的项

```python
res=pd.merge(left,right,on=['key1','key2'],how='outer')
res
```

左外连接，左边未匹配全部保留

```python
res=pd.merge(left,right,on=['key1','key2'],how='left')
res
```

右外连接，右边未匹配全部保留

```python
res=pd.merge(left,right,on=['key1','key2'],how='right')
res
```

显示每一行的连接方式：两边都有（both）,只有左边（left_only），只有右边（right_only）


```python
res=pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)
res
```

给indecator指定名称

```python
res=pd.merge(left,right,on=['key1','key2'],how='outer',indicator='Indecator')
res
```

用index作为连接

```python
left=pd.DataFrame({'A':['A0','A1','A2'],
                  'B':['B0','B1','B2']},
                 index=['K0','K1','K2'])
right=pd.DataFrame({'C':['C0','C1','C2'],
                  'D':['D0','D1','D2']},
                 index=['K0','K2','K3'])
print(left)
print(right)
```

left_index和right_index默认为False

```python
res=pd.merge(left,right,left_index=True,right_index=True) #默认内连接（how='inner'）
res
```

同样可以外链接

```python
res=pd.merge(left,right,left_index=True,right_index=True,how='outer') #默认内连接（how='inner'）
res
```

更名suffixes(后缀)
两个表中有相同的列（属性），但是表达不同的含义，为了区分我们可以给他们加后缀
k表示员工编号（男女有重的，女跟女也有重的），salary表示工资

```python
boys=pd.DataFrame({'k':['K0','K1','K2'],
                  'salary':[1,2,3]})
girls=pd.DataFrame({'k':['K0','K0','K3'],
                  'salary':[4,5,6]})
print(boys)
print(girls)
```

```python
res=pd.merge(boys,girls,on='k',how='outer')
res
```

```python
res=pd.merge(boys,girls,on='k',suffixes=['_boy','_girl'],how='outer')
res
```

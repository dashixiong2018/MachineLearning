```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Series
data=pd.Series(np.random.randn(1000))
data=data.cumsum()                  #之前的数据全部累加
data.plot()
plt.show()
```

```python
#DataFrame
data=pd.DataFrame(np.random.randn(1000,4),columns=list("ABCD"))
data=data.cumsum()
print(data.head())          #默认显示前四行可以加参数设置显示行数
data.plot()                 #plot()中可以添加参数设置颜色、线型线宽等                      
plt.show()
```

```python
#以上都是线型的，当要绘制离散的点时用scatter（散布）
#只能是二维的点，所以必须设置x和y
ax=data.plot.scatter(x='A',y='B',color='Blue',label='Class 1')
data.plot.scatter(x='A',y='C',color='Red',label='Class 2',ax=ax)
#多一个参数ax=ax，代表将当前行的点嵌入到ax中
plt.show()
```
